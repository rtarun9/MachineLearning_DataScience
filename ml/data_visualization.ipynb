```python
import seaborn as sb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the tips data set
tips = sb.load_dataset('tips')
tips.head()
```

```python
# Visualize the plot
sb.relplot(data=tips, x='total_bill', y='tip', hue='smoker', size='size', col='time')
```

```python
# Load iris data set from seaborn
iris = sb.load_dataset('iris')
iris.head()
```

```python
sb.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', size='petal_width')
```

```python
# Load the titanic data set
titanic = sb.load_dataset('titanic')
titanic.head()
```

```python
sb.countplot(data=titanic, x='class', hue='sex')
```

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

house = pd.DataFrame(housing.data, columns=housing.feature_names)
house['price'] = housing.target
house.head()
```

```python
# Distribution plot
sb.displot(data=house, x='price', kde=True)
```

```python
# Heat map for correlation

corr = titanic.corr()

sb.heatmap(data=corr, annot=True)
```
