```python
# Calculate entropy of whole data set (s)
# s can be either 0 or 1.
zeros_count = 5
yes_count = 9

# Calculate information gain of each and every attribute
entropy_s = -1.0 * 9/14 * np.log2(9/14) + (-1.0 * 5/14 * np.log2(5/14))
print("Entropy(S) : ", entropy_s)

# Calculate entropy for feature 1 : outlook.
# sunny has 3 0's and 2 1's.
entropy_sunny = -1.0 * 2/5 * np.log2(2/5) + (-1.0 * 3/5 * np.log2(3/5))

# overcast has 4 1's and 0 0's.
entropy_overcast = -1.0 * 4/4 * np.log2(4/4) 

# rain has 3 1's and 2 0's.
entropy_rain = -1.0 * 3/5 * np.log2(3/5) + (-1.0 * 2/5 * np.log2(2/5))

entropy_outlook =  5/14.0 * entropy_sunny + 4/14.0 * entropy_overcast + 5/14.0 * entropy_rain

print("Entropy outlook : ", entropy_outlook)
ig_outlook = entropy_s - entropy_outlook
print("IG : outlook : ", ig_outlook)
```
