```python
# Speciailized numpy arrays
a = np.zeros((2,2))
print(a)

b = np.ones((2,2))
print(b)

c = np.full((2, 2), 199)
print(c)

d = np.eye(2)
print(d)

e = np.random.random((2,2))
```

```python
# Array math.
x = np.array([[1,2,3],[4,5,6]], dtype=np.float32)
y = np.array([[3,3,3],[5,5,5]])
print(x+y)
print(np.add(x,y))

print(np.multiply(x,y))
print(np.sqrt(x))
```

```python
# Statistics from numpy array.
a = np.random.randn(3,3)
print(a.mean())
print(a.std())
print(a.sum())
```

```python
# Using matplotlib for data visualization.
import matplotlib.pyplot as plt
import numpy as np

# equally spaced data between 0 and 10
x = np.linspace(0, 10, 100)
print(x)

# sin(x) and cos(x)
y = np.sin(x)
z = np.cos(x)

# plot x vs sin(x) and x vs cos(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('x  vs sin(x)')
plt.show()

plt.plot(x, z)
plt.title('x vs cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.show()

# bar plot
x = ['English', 'French', 'Spanish', 'Latin', 'German']
y = [20, 40, 20, 60, 80]
plt.bar(x, y)
plt.xlabel('languages')
plt.ylabel('population (in m)')
plt.title('language vs population')
plt.show()
```
