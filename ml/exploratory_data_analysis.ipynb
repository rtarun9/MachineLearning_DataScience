```python
import numpy as np
import pandas as pd
import statistics as st

# Finding mean of some data manually and by using numpy
sample_list = [1, 2, 2, 2, 4, 5, 6, 6, 6, 9, 10, 10, 10, 11, 14, 14, 15, 18, 19, 20]

mean = 0
for i in sample_list:
    mean += i

mean /= len(sample_list)
print("Mean :", mean)

np_list = np.asarray(sample_list)
print("Mean :", np.mean(np_list))

print("Median :", np.median(np_list))

print("Mode :", st.mode(np_list))

char_arr = ['a', 'c', 'c', 'x']
print("Mode :", st.mode(char_arr))
```

```python
# Variance and standard deviation
data = [1, 3, 3, 3, 2, 4, 5, 6, 34, 2, 1, 5, 6]
data = np.asarray(data)

print("Variation: ", np.var(data))
print("Standard deviation: ", np.std(data))
```

```python
# Data visualization
import seaborn as sb
import matplotlib.pyplot as plt

# Create a random array (following normal distribution)
x = np.random.normal(size=(30, 30))
# print(x, x.shape)

# General statistical information from the data
print("Max :", np.max(x))
print("Min :", np.min(x))

# Plotting frequency histogram using matplotlib
plt.hist(x)
plt.xlabel('Intervals')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot the same histogram using seaborn
sb.histplot(x, bins=5)
```

```python
sb.histplot(x, bins=10)
```

```python
# Using density plots
sb.displot(x, kind="kde")

# Plot the mean and median
plt.axvline(x.mean(), color='g')
plt.axvline(np.median(x), color='purple')

plt.show()
```

```python
# Quartiles
q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)
print("IQR :", q3 -q1)
```

```python
# Box plot
sb.boxplot(x)
```

```python
 # Detecting outliers using IQR
q1 = np.percentile(x, 25)
q3 = np.percentile(x, 75)

iqr = q3 - q1
lower_bound = (q1) - iqr * 1.5
upper_bound = q1 + iqr * 1.5

x = np.random.random(size=(1,10))

for i in x[0]:
    if (i > upper_bound) or (i < lower_bound):
        print("outlier detected:", i)
```

```python
# Using Z score to remove outliers
from scipy import stats

x = np.random.rand(50, 100)
z = np.abs(stats.zscore(x))

# Using 0.5 as a threshold value
x2 = x[(z < 0.5).all(axis=1)]
print(x.shape, x2.shape)
```
