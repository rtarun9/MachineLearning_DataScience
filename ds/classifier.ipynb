Question 5. Perpare a classifier model for wine data set

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
import pandas as pd

from sklearn import neighbors
from sklearn import naive_bayes
from sklearn.linear_model import Perceptron

import scipy.stats as stats
import math
```

```python
df = pd.read_csv('../data_sets/wine.csv')
df
```

```python
# Use KNN classifier

x = df.iloc[:,1:-1].values
y = df.iloc[:,0]
print("Target : \n", y)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from matplotlib import pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4000)

from sklearn.neighbors import KNeighborsClassifier
cl = KNeighborsClassifier(n_neighbors=5, p=3)
cl.fit(x_train, y_train)

y_pred = cl.predict(x_test)
print('Y prediction : ', y_pred)

print("Y test", y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix : ", cm)

from sklearn.metrics import accuracy_score
print("accuracy_score : ", accuracy_score(y_test, y_pred))
```

```python
# Use Logistic regression classifier.

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)

scaler = preprocessing.StandardScaler().fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = LogisticRegression()
model.fit(x_train_scaled, y_train)

train_acc = model.score(x_train_scaled, y_train)
print("Accuracy for training set : ", train_acc * 100)

y_pred = model.predict(x_test_scaled)
test_acc = accuracy_score(y_test, y_pred)
print("Accuracy for test set : ", test_acc * 100)

print(classification_report(y_test, y_pred))

c_mat = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", c_mat)

print("accuracy_score : ", accuracy_score(y_test, y_pred))
```

Use Perceptron Classifier

```python
p_classifier = Perceptron(tol=1e-3, random_state=1)
p_classifier.fit(x, y)
print("Score using Perceptron classifier : ", p_classifier.score(x, y))
```
