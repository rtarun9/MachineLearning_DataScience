```python
median_compensation = df['Total Compensation'].quantile(0.5)
print("Median compensation : ", median_compensation)
```

```python
# Randomly sampling 500 employees at random and observe the median compensation.
samples = df['Total Compensation'].sample(500)
samples
```

```python
samples.hist()
median_sample_compensation = samples.quantile(0.5)
plt.scatter(x=median_sample_compensation, y=0, color='red', s=50)
```

```python
# It can be observed that the sample mean is very close to the sample mean.
# However, this might be because of chance, so we have to quantify the amound by which 
# this estimate could vary among various samples.
```

```python
# Bootstrapping
smp = df['Total Compensation'].sample(500, replace=True)
smp
```

```python
smp.quantile(0.50)
```

```python
smp.hist()
plt.scatter(smp.quantile(0.50), 0, color='red')
```

```python
# Here, the results are similar to the previously obtained value (i.e median is fairly
# close to the sample mean) This is another estimate that we have obtained).
```

```python
# Bootstrap function : return array of sample medians
def bootstrap_median(sample, replications):
    smp = []

    for i in range(replications):
        s = sample.sample(500, replace=True)
        smp.append(s.quantile(0.50))
    
    return smp
```

```python
bootstrapped_medians = bootstrap_median(df['Total Compensation'], 1000)
bootstrapped_df = pd.DataFrame(bootstrapped_medians)
bootstrapped_df.head(10)
```

```python
bootstrapped_df.hist()
plt.scatter(x=df['Total Compensation'].quantile(0.50), y=0, color='red')
```

```python
# It can be seen that even after 1000 random samples, the population median is similar
# to the sampled medians (of around 110,000.). This means that there is no 'chance'
# involved in the population median, as the randomly sampled (bootstrapped) samples are very close to the 
# population median.
```
