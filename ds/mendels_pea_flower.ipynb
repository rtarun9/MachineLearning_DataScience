```python
import numpy as np
```

```python
flowers = ['purple'] * 705
white = ['white'] * 224

flowers.extend(white)
np.random.shuffle(flowers)
flowers
```

```python
# Find number of purple flowers
purple_count = 0
for f in flowers:
    if f == 'purple':
        purple_count += 1

print(purple_count / np.size(flowers))
```

# We will take test statistic to be p(purple) = 0.75 

```python
observed_statistic = abs((purple_count / np.size(flowers)) - 0.75)
observed_statistic
```

```python
# Simulation
n = 1000

results = []
for i in range(n):
    results.append(np.random.binomial(np.size(flowers), 0.75) / np.size(flowers))

print(results)
# Distance from 0.75
dist = []
for i in results:
    dist.append(np.abs(i - 0.75))

```

```python
# Visualize the distances
import seaborn as sb

sb.histplot(dist)
```
